```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('data.csv')
```

```python
dataset
```

```python
print(dataset.describe())
```

```python
# Find Outlier

plt.figure(figsize=(10,10))
dataset.boxplot()
plt.show()
```

```python
# Percentile Batas Bawah

lowwer_limit = dataset['Popularity'].quantile(0)
lowwer_limit
```

```python
# Menampilkan data lowwer limit

dataset[dataset['Popularity'] > lowwer_limit]
```

```python
# Persentil batas atas

upper_limit = dataset['Popularity'].quantile(1)
upper_limit
```

```python
dataset[dataset['Popularity'] < upper_limit]
```

```python
# Remove Outlier

dataset_no_outlier = dataset[(dataset['Popularity'] > lowwer_limit) & (dataset['Popularity']< upper_limit)]
```

```python
dataset_no_outlier
```

```python
plt.figure(figsize=(10,10))
dataset_no_outlier.boxplot()
plt.show()
```

```python
# Cek jumlah outlier yang terhapus

dataset.shape[0] - dataset_no_outlier.shape[0]
```
